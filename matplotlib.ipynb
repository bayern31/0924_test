```python
import matplotlib.pyplot as plt

# Data
dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(dates, min_temperature, label='Min Temperature', marker='o', linestyle='-', color='b')
plt.plot(dates, max_temperature, label='Max Temperature', marker='o', linestyle='-', color='r')

# Formatting the plot
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Daily Min and Max Temperatures')
plt.xticks(rotation=45)
plt.legend()
plt.grid(visible=True)

# Display plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]

# Creating figure and axis objects using the object-oriented API
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting data on the axis
ax.plot(dates, min_temperature, label='Min Temperature', marker='o', linestyle='-', color='b')
ax.plot(dates, max_temperature, label='Max Temperature', marker='o', linestyle='-', color='r')

# Formatting the plot
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (C)')
ax.set_title('Daily Min and Max Temperatures')
ax.tick_params(axis='x', rotation=45)
ax.legend()
ax.grid(visible=True)

# Adjust layout and show plot
fig.tight_layout()
plt.show()
```

# Yahoo API    

```python
import yfinance as yf
import matplotlib.pyplot as plt

#  
data = yf.download("AAPL", start="2020-01-01", end="2024-09-30")

#     
data2 = data.reset_index()

#  
fig, ax = plt.subplots()

# x    'Date'   
ax.plot(data2['Date'], data2['Open'], label='Open Price')

#   
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.set_title('AAPL Open Prices')
ax.legend()

#  
plt.show()
```

```python
import yfinance as yf
apple = yf.download("AAPL", start="2020-01-01", end="2024-09-30")
nvidia = yf.download("NVDA", start="2020-01-01", end="2024-09-30")
intel = yf.download("INTC", start="2020-01-01", end="2024-09-30")

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(10, 4))

ax[0, 0].plot(apple['Open'], label = "Apple")
ax[0, 1].plot(nvidia['Open'], label = "Nvidia")
ax[1, 0].plot(intel['Open'], label = "Intel")

ax[0, 0].set_title("Apple stocks")
ax[0, 1].set_title("Nvidia stocks")
ax[1, 0].set_title("Intel stocks")

ax[0, 0].legend()
ax[0, 1].legend()
fig.tight_layout()
plt.savefig("myStocks.png")
plt.show()
```
