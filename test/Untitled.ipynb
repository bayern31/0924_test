```python
import pandas as pd

# 데이터프레임 생성
data = [
    ["1,000", "1,100", '1,510'],
    ["1,410", "1,420", '1,790'],
    ["850", "900", '1,185'],
]
columns = ["03/02", "03/03", "03/04"]
df = pd.DataFrame(data=data, columns=columns)
df.info()
```

```python
# 쉼표 제거하고 int로 변환하는 함수 정의

x = "1,000" # 처음에 이정도부터 바꾸는 연습 ㄱㄱ
int(x.replace(",", ""))
```

```python
def rm_comma(x):
    return int(x.replace(",", ""))

y = '1,000'
print(rm_comma(y))
```

```python
df["03/02"] = df["03/02"].apply(rm_comma)
df["03/03"] = df["03/03"].apply(rm_comma)
df.info()
```
